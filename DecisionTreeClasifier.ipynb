# Decision Tree Classifier


```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("C:\\Users\\Abc\\Downloads\\covid_toy.csv")
```

```python
df=df.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df['gender']=lb.fit_transform(df['gender'])
df['cough']=lb.fit_transform(df['cough'])
df['city']=lb.fit_transform(df['city'])
df['has_covid']=lb.fit_transform(df['has_covid'])
```

```python
df.head()
```

```python
x=df.drop(columns =['has_covid'])
y=df['has_covid']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2 , random_state=42)
```

# logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train , y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test , y_pred)
```

# DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt= DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test ,y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid ={
    'max_depth' :[10,20,30,None],
    'min_samples_split':[2,5,10],
    'min_samples_leaf' :[1,2,4]
}
```

```python
dtree_clf = DecisionTreeClassifier(random_state=42)
```

```python
grid_search= GridSearchCV(estimator=dtree_clf , param_grid=param_grid,
                          cv=5 , n_jobs=-1 , verbose=2 , scoring='neg_mean_squared_error')
```

```python
grid_search.fit(x_train , y_train)
best_dtree_reg = grid_search.best_estimator_
```

```python
best_dtree_reg
```

```python
y_pred = best_dtree_reg.predict(x_test)
```

```python
ac=accuracy_score(y_test , y_pred)
```

```python
best_params = grid_search.best_params_
print(f"Best parameters:{best_params}" )
```
